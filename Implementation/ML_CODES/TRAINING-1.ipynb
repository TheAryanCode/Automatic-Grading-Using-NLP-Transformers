```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('C:\\Users\\V R N S Nikhil\\OneDrive\\Desktop\\4th_sem\\ML\\FINAL\\ML\\Auto.xlsx')
df.head()
```

```python
x=df.drop('output',axis=1)
y=df['output']
```

```python
def custom_round(value):
    integer_part = int(value)
    decimal_part = value - integer_part
    if decimal_part == 0.5:
        return value
    else:
        return round(value)

df['output'] = df['output'].apply(custom_round)
df['output'].value_counts()
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
base=KNeighborsRegressor()
```

```python
help(KNeighborsRegressor)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
scale_x_train=std.fit_transform(x_train)
scale_x_test=std.transform(x_test)
```

```python
scale_x_train
```

```python
param_grid={'n_neighbors':list(np.arange(100)),
            'weights':['uniform','distance'],
             'metric':['minkowski','euclidean','chebyshev']}
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
gtkk=GridSearchCV(base,param_grid)
```

```python
gtkk.fit(scale_x_train,y_train)
```

```python
gtkk.best_score_
```

```python
test = gtkk.predict(scale_x_test)
```

```python
test
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```
